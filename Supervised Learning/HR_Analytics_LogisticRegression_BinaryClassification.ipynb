From the data analysis we conclude that we will use following variables as independant variables in our model.

1) **Satisfaction Level**

2) **Average Monthly Hours**

3) **Promotion Last 5 Years**

4) **Salary**


```python
Indep_df = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
Indep_df.head()
```

```python
#Convert the salary variable to numbers for that we need to use dummy variable for that which can be got from pandas get_dummies function
salary_dummies = pd.get_dummies(Indep_df.salary, prefix ="salary")
salary_dummies
```

```python
#Concatinate the salary_dummies with the df we have to use. \
# Also we need to remove the text part of the salary variable as now we have the dummies for salary
df_withDummies = pd.concat([Indep_df, salary_dummies], axis = "columns")
df_withDummies
```

```python
df_withDummies.drop(['salary'], axis='columns', inplace=True)
df_withDummies.head()
```

```python
X = df_withDummies
X
```

```python
Y = df.left
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.9)

```

```python
from sklearn.linear_model import LogisticRegression
analytics_model = LogisticRegression()
```

```python
analytics_model.fit(X_train, Y_train)
```

```python
analytics_model.predict(X_test)
```

Testing the Accuracy of our model

```python
analytics_model.score(X_test, Y_test)
```

We noticed that the accuracy is just 78.8% which is low. It means that there are other factors that are affecting the people leaving the company. Maybe by taking the Department wise consideration, we will have a better accuracy. That task is for later.
